The purpose of this Notebook is to analyze decades-long trends in petroleum field energy return on investment (EROI).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax
```

```python
df = pd.read_excel("Energy Return on Investment.xlsx")
xls = pd.ExcelFile('Energy Return on Investment.xlsx')
```

```python
xls.sheet_names
```

```python
#Read the weights spreadsheets
df_cantarell_weights = pd.read_excel('Energy Return on Investment.xlsx', 'Cantarell Weights')
df_forties_weights = pd.read_excel('Energy Return on Investment.xlsx', 'Forties Weights')
df_midway_sunset_weights = pd.read_excel('Energy Return on Investment.xlsx', 'Midway-Sunset Weights')
df_prudhoe_bay_weights = pd.read_excel('Energy Return on Investment.xlsx', 'Prudhoe Bay Weights')
df_wilmington_weights = pd.read_excel('Energy Return on Investment.xlsx', 'Wilmington Weights')
```

```python
# Analyze the Cantarell Weights Spreadsheet
year_cantarell = df_cantarell_weights.iloc[:,0]
free_flow_share_cantarell = df_cantarell_weights.iloc[:,1]
gas_lift_share_cantarell = df_cantarell_weights.iloc[:,2]
plt.plot(year_cantarell, free_flow_share_cantarell, label = "Free Flow Share")
plt.plot(year_cantarell, gas_lift_share_cantarell, label = "Gas Lift Share")
plt.xlabel("Year")
plt.ylabel("Share of Production")
plt.title("Share of Production in Cantarell")
plt.legend()
```

```python
# Analyze the Forties Weights Spreadsheet
year_forties = df_forties_weights.iloc[:,0]
free_flow_share_forties = df_forties_weights.iloc[:,1]
gas_lift_share_forties = df_forties_weights.iloc[:,2]
downhole_pump_share_forties = df_forties_weights.iloc[:,3]
plt.plot(year_forties, free_flow_share_forties, label = "Free Flow Share")
plt.plot(year_forties, gas_lift_share_forties, label = "Gas Lift Share")
plt.plot(year_forties, downhole_pump_share_forties, label = "Downhole Pump Share")
plt.xlabel("Year")
plt.ylabel("Share of Production")
plt.title("Share of Production in Forties Field")
plt.legend()
```

```python
# Analyze the Midway Sunset Weights Spreadsheet
year_midway_sunset = df_midway_sunset_weights.iloc[:,0]
steam_from_crude_oil_midway_sunset = df_midway_sunset_weights.iloc[:,1]
steam_from_natural_gas_midway_sunset = df_midway_sunset_weights.iloc[:,2]
steam_from_natural_gas_from_OTSG_midway_sunset = df_midway_sunset_weights.iloc[:,3]
steam_from_natural_gas_from_HRSG_midway_sunset = df_midway_sunset_weights.iloc[:,4]
total_steam_from_OTSG = steam_from_natural_gas_midway_sunset * steam_from_natural_gas_from_OTSG_midway_sunset
total_steam_from_HRSG = steam_from_natural_gas_midway_sunset * steam_from_natural_gas_from_HRSG_midway_sunset
plt.plot(year_midway_sunset,steam_from_crude_oil_midway_sunset, label = "Crude Oil" )
plt.plot(year_midway_sunset, total_steam_from_OTSG, label = "OTSG")
plt.plot(year_midway_sunset, total_steam_from_HRSG, label = 'HRSG')
plt.xlabel("Year")
plt.ylabel("Share of Production")
plt.title("Share of Production in Midway Sunset Field")
plt.legend()
```

```python
# Analyze the Purdhoe Bay Weights Spreadsheet
year_prudhoe_bay = df_prudhoe_bay_weights.iloc[:,0]
free_flow_share_prudhoe_bay = df_prudhoe_bay_weights.iloc[:,1]
gas_lift_share_prudhoe_bay = df_prudhoe_bay_weights.iloc[:,2]
plt.plot(year_prudhoe_bay, free_flow_share_prudhoe_bay, label = "Free Flow Share")
plt.plot(year_prudhoe_bay, gas_lift_share_prudhoe_bay, label = "Gas Lift Share")
plt.xlabel("Year")
plt.ylabel("Share of Production")
plt.title("Share of Production in Prudhoe Bay Field")
plt.legend()
```

```python
# Analyze the Wilmington Weights Spreadsheet
year_wilmington = df_wilmington_weights.iloc[:,0]
steam_from_crude_oil_wilmington = df_wilmington_weights.iloc[:,1]
steam_from_natural_gas_wilmington = df_wilmington_weights.iloc[:,2]
steam_from_natural_gas_from_OTSG_wilmington = df_wilmington_weights.iloc[:,3]
steam_from_natural_gas_from_HRSG_wilmington = df_wilmington_weights.iloc[:,4]
total_steam_from_OTSG_wilmington = steam_from_natural_gas_wilmington * steam_from_natural_gas_from_OTSG_wilmington
total_steam_from_HRSG_wilmington = steam_from_natural_gas_wilmington * steam_from_natural_gas_from_HRSG_wilmington
plt.plot(year_wilmington,steam_from_crude_oil_wilmington, label = "Crude Oil" )
plt.plot(year_wilmington, total_steam_from_OTSG_wilmington, label = "OTSG")
plt.plot(year_wilmington, total_steam_from_HRSG_wilmington, label = 'HRSG')
plt.xlabel("Year")
plt.ylabel("Share of Production")
plt.title("Share of Production in Wilmington Field")
plt.legend()
```

The following code shows the change in free flow share as the fields of interest age. 
The fields of interest are the Prudhoe Bay Field, the Forties Field, and the Cantarell Field

```python
plt.figure(figsize = (8,5))
new_year_cantarell = year_cantarell - year_cantarell[0]
new_year_prudhoe_bay = year_prudhoe_bay - year_prudhoe_bay[0]
new_year_forties = year_forties - year_forties[0]
plt.plot(new_year_cantarell,free_flow_share_cantarell, label = "Cantarell", c = 'b', linewidth = 6.0)
plt.plot(new_year_prudhoe_bay, free_flow_share_prudhoe_bay,label = "Prudhoe Bay", c = 'r', linewidth = 6.0)
plt.plot(new_year_forties, free_flow_share_forties, label = "Forties", c = 'g', linewidth = 6.0)
plt.xlabel("Years of Operation", fontsize = 14)
plt.ylabel("Share of Production that is Free Flow", fontsize = 14)
plt.title("Production Share that is Free Flow vs Years of Operations", fontsize = 16)
plt.legend(fontsize =13)
```
